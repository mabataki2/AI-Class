```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score
import pandas as pd


# 1. 데이터 준비
df = pd.read_csv("/content/drive/MyDrive/winequality-white.csv", sep=';').dropna()
X = df.drop('quality', axis=1)
y = df['quality']

df
```

```python
# 레이블 인코더를 사용하여 quality 열을 0부터 시작하는 값으로 변환
le = LabelEncoder()
y_encoded = le.fit_transform(y)

```

```python

# 학습 및 테스트 데이터 분할
# y_encoded를 사용하여 데이터를 분할합니다.
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python

# KNN과 로지스틱 회귀를 위한 데이터 스케일링
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python

# 2. 모델 학습 및 평가
print("--- KNN Classifier ---")
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)
knn_y_pred = knn.predict(X_test_scaled)
knn_accuracy = accuracy_score(y_test, knn_y_pred)
print(f"Accuracy: {knn_accuracy:.4f}\n")

print("--- Linear Regression ---")
lr_reg = LinearRegression()
lr_reg.fit(X_train_reg, y_train_reg)
lr_reg_y_pred = lr_reg.predict(X_test_reg)
lr_reg_mse = mean_squared_error(y_test_reg, lr_reg_y_pred)
lr_reg_r2 = r2_score(y_test_reg, lr_reg_y_pred)
print(f"Mean Squared Error: {lr_reg_mse:.4f}")
print(f"R-squared: {lr_reg_r2:.4f}")
# 절편과 기울기 출력
print("\n--- Linear Regression Coefficients & Intercept ---")
print(f"Intercept: {lr_reg.intercept_:.4f}")
coefficients = pd.DataFrame(lr_reg.coef_, X.columns, columns=['Coefficient'])
print(coefficients)
print("\n")
```
